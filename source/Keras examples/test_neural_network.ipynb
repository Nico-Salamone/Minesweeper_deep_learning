```python
from random import randint

from keras.models import Sequential
from keras.layers import Dense
from keras.models import load_model
```

```python
def count_ones(x):
    count = 0
    for i in range(len(x)):
        if x[i] == 1:
            count += 1
    
    return count

def one_ratio(x):
    n = len(x)
    count = count_ones(x)
    
    return count / n
```

```python
data_set_size = 5000
n = 10

x = [
[randint(0, 1) for j in range(n)]
for i in range(data_set_size)
]

y_true = []
for i in range(data_set_size):
    #y_true.append(count_ones(x[i]))
    y_true.append(one_ratio(x[i]))

"""
for i in range(data_set_size):
    print(x[i])
    print(y_true[i])
    print()
"""
```

```python
model = Sequential()
model.add(Dense(n, input_dim=n, activation='relu'))
model.add(Dense(30))
model.add(Dense(1, activation='relu'))

model.compile(loss = 'mean_squared_error', optimizer = 'rmsprop', metrics = ['mean_squared_error', 'mean_absolute_error'])

model.fit(x, y_true, epochs=10, batch_size=1)
```

```python
scores = model.evaluate(x, y_true, verbose=0)
for i in range(len(scores)):
    print("{}: {:.3f}".format(model.metrics_names[i], scores[i]))
```

```python
y_pred = model.predict(x)
for i in range(data_set_size):
    print(x[i])
    print(y_pred[i])
    print()
```
